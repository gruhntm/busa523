```python
# 1
# Write a program which randomly selects a number between 1 and 20. The user is given 5 chances to guess the number
# correctly to win.

import random

num = random.randint(1,20)
print('Please enter a number between 1 and 20')
input_num = input()
input_num = int(input_num)
if input_num == num:
    print("You guessed it!")
else:
    print('Incorrect. Please try again.')
    for x in range(0,4):
        print('Please enter a number between 1 and 20')
        input_num = input()
        input_num = int(input_num)
        if input_num == num:
            print("You guessed it!")
            break
        else:
            print('Incorrect. Please try again.')
            ++x
```
